```python
from selenium import webdriver
from fake_useragent import UserAgent
import time
```

```python
movieName = input("영화제목을 입력해주세요 : ")
```

```python
rows = []
def crawlingMovie(headless=False):
    options = webdriver.ChromeOptions()
    options.add_argument("user-agent={}".format(UserAgent().chrome))
    
    if headless:
        options.add_argument("headless")
    
    
    driver = webdriver.Chrome(options=options)
    driver.get("https://movie.naver.com/movie/running/current.nhn")
    # 영화 검색어 입력
    driver.find_element_by_css_selector("#ipt_tx_srch").send_keys(movieName)
    # 영화 검색버튼 클릭
    driver.find_element_by_css_selector("#jSearchArea div button").click()
    time.sleep(1)
    # 영화 검색 결과중 1번째 li 클릭
    driver.find_element_by_css_selector("#cbody > #old_content ul.search_list_1 > li > p > a").click()
    time.sleep(1)
    # 댓글 화면으로 들어가기
    driver.find_element_by_css_selector("div.score.score_left > div.star_score  > a").click()
    time.sleep(2)
    
    for i in range(1,10+1):
        
        driver.execute_script("window.scrollTo(0, 2500);")
        if i == 1 :
            iframe = driver.find_element_by_css_selector("#pointAfterListIframe")
            driver.switch_to_frame(iframe)

        datguls = driver.find_elements_by_css_selector("div.input_netizen  > div.score_result > ul li")

        # 평점,댓글,공감수,비공감수 크롤링
        for datgul in datguls:
            row = []
            score = datgul.find_element_by_css_selector("div.star_score > em").text
            text = datgul.find_element_by_css_selector("div.score_reple > p").text
            gong = datgul.find_element_by_css_selector("div.btn_area > strong:nth-child(2) > span").text
            bigong = datgul.find_element_by_css_selector("div.btn_area > strong:nth-child(4) > span").text
            row.append(score)
            row.append(text)
            row.append(gong)
            row.append(bigong)
            rows.append(row)
            
        driver.find_element_by_css_selector("#pagerTagAnchor{} > em".format(i+1)).click()
    
    columns = ["평점", "댓글", "공감수", "비공감수"]
    return pd.DataFrame(rows, columns=columns)

#testcode
df = crawlingMovie()
```

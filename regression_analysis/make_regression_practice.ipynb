### make_regression : 1 feature 

#### 연습문제 1
- make_regression과 같은 기능을 하는 함수 make_regression2를 만들어라. 단 make_regression2는 coef=True, n_features=1 라고 가정한다. 즉 항상 가중치 계수를 반환하고 1차원 독립 변수만 생성할 수 있다. 따라서 make_regression2는 다음과 같은 인수만 가진다.

    - `n_samples`
    - `bias`
    - `noise`
    - `random_state`
    

```python
def make_regression2(n_samples, bias, noise, random_state):
    np.random.seed(random_state)
    
    x = np.random.randn(1,n_samples)
    noise = sp.stats.norm(0, noise).rvs(n_samples)
    w = np.random.randn()*100
    y = bias + x*w + noise
    
    return x, y, w
```

```python
x,y,w = make_regression2(50, 3, 40, 0)
```

### make_regression : 2 feature 

```python
X, y, w = make_regression(n_samples=300, n_features=2, noise=0, coef=True, random_state=0)

plt.scatter(X[:, 0], X[:, 1], c=y, s=100, cmap=mpl.cm.bone)
plt.xlabel("x1")
plt.ylabel("x2")
plt.axis("equal")
plt.title("two independent variables are correlated with a dependent variable")
plt.show()
```

```python
X, y, w = make_regression(n_samples=300, n_features=2, n_informative=1,
                          noise=0, coef=True, random_state=0)

plt.scatter(X[:, 0], X[:, 1], c=y, s=100, cmap=mpl.cm.bone)
plt.xlabel("x1")
plt.ylabel("x2")
plt.axis("equal")
plt.title("only one of independent variable is correlated with a dependent variable")
plt.show()
```

```python
X, y, w = make_regression(n_samples=300, n_features=2, effective_rank=1, noise=0,
                          tail_strength=0, coef=True, random_state=0)

plt.scatter(X[:, 0], X[:, 1], c=y, s=100, cmap=mpl.cm.bone)
plt.xlabel("x1")
plt.ylabel("x2")
plt.axis("equal")
plt.title("independent variables are not independent of each other")
plt.show()
```

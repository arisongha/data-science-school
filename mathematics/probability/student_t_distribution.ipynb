### 스튜던트 t 분포

#### 1. 팻-테일(Fat tail) 현상
- 분포의 모양을 볼 때 양 끝(꼬리) 부분이 정규 분포보다 두껍다고 해서 이를 Fat tail 현상이라고 한다.

```python
%matplotlib inline
```

```python
xx = np.linspace(-4, 4, 100)
for df in [1,2,5,10,20]:
    rv = sp.stats.t(df=df)
    plt.plot(xx, rv.pdf(xx), label=("student-t (dof=%d)"% df))
plt.plot(xx, sp.stats.norm().pdf(xx), label="Normal", lw=5, alpha=0.5)
plt.title("chage of student-t distribution through degree of freedom")
plt.xlabel("sample")
plt.ylabel("p(x)")
plt.legend()
plt.show()
```

#### 2. 통계량 분포
- 가우시안 정규 분포로부터 얻은 N개의 표본 x1,...,xN에서 계산한 표본평균을 표본표준편차로 정규화한 값을 t통계량이라고 한다

```python
np.random.seed(0)

rv = sp.stats.norm()
M = 1000

plt.subplot(1,2,1)
N = 4
x1 = rv.rvs((N,M))
xbar1 = x1.mean(axis=0)
xstd1 = x1.std(axis=0, ddof=1)
x = xbar1/(xstd1/np.sqrt(N))
sns.distplot(x, kde=True)
xx = np.linspace(-6,6,1000)
plt.plot(xx, rv.pdf(xx), 'r:', label="normal")
plt.xlim(-6,6)
plt.ylim(0,0.5)
plt.title("t distribution (N=4)")
plt.legend()

plt.subplot(1,2,2)
N = 40
x2 = rv.rvs((N,M))
xbar2 = x2.mean(axis=0)
xstd2 = x2.std(axis=0, ddof=1)
x = xbar2 / (xstd2/np.sqrt(N))
sns.distplot(x, kde=True)
xx = np.linspace(-6,6,1000)
plt.plot(xx, rv.pdf(xx), 'r:', label="normal")
plt.xlim(-6,6)
plt.ylim(0,0.5)
plt.title("t distribution (N=4)")
plt.legend()

plt.show()
```

### 카이 제곱 분포
- 가우시안 정규 분포를 따르는 확률변수 X의 N개의 샘플들을 제곱을 하여 더하면 양수값만 가지는 분포가 된다.

```python
%matplotlib inline
```

```python
xx = np.linspace(0.01, 10, 100)
dfs = np.arange(1, 5)
lss = ["-", "--", "-.", ":"]
for df ,ls in zip(dfs, lss):
    rv = sp.stats.chi2(df=df)
    plt.plot(xx, rv.pdf(xx), ls=ls, label = ("chi2 (dof=%d)" % df))
plt.xlim(0, 10.1)
plt.ylim(0, 0.6)
plt.title("chage of chi-squared distribution through free")
plt.xlabel("sample")
plt.ylabel("p(x)")
plt.legend()
plt.show()
```

```python
np.random.seed(0)

rv = sp.stats.norm()
M = 2000

plt.subplot(1,2,1)
N = 2
x  = rv.rvs((N,M))
t = (x ** 2).sum(axis=0)
sns.distplot(t, kde=False)
plt.xlim(-1,10)
plt.ylim(0,800)
plt.title("distribution of squared-sum(N=2")
plt.xlabel("sample")
plt.ylabel("count")

plt.subplot(1,2,2)
N = 30
x  = rv.rvs((N,M))
t = (x ** 2).sum(axis=0)
sns.distplot(t, kde=False)
plt.xlim(-1,60)
plt.ylim(0,800)
plt.title("distribution of squared-sum(N=30)")
plt.xlabel("sample")
plt.ylabel("count")

plt.show()
```
